Membuat variabel baru dan memasukkan data S11 dan S21 kedalam array

```python
#ambil data dari tiap kolom
x_s11 = data.frekuensi
s11real = data.S11real
s11im = data.S11im
s21real= data.S21real
s21im = data.S21im
```

```python
#ubah frame ke dalam array
frekuensi = np.array(x_s11)
s11re = np.array(s11real)
s11im = np.array(s11im)
s21re = np.array(s21real)
s21im = np.array(s21im)
```

```python
#ubah nilai real dan imaginer kedalam bentuk kompleks
s11=s11re + 1j*s11im
s21=s21re + 1j*s21im
```

**MENGHITUNG NILAI EPSILON RELATIF DAN MIU RELATIF**

**Menghitung nilai impedansi gelombang dari nilai array S11 dan S21**

```python
#z=impedansi gelombang

a=(((1+s11)**2)-(s21**2))
b=(((1-s11)**2)-(s21**2))
z= (a/b)**0.5
z
```

```python
zre=z.real
zre
```

```python
zim=z.imag
zim
```
